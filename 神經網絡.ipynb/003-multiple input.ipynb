```python
import csv
reader=csv.reader(open("grades-multiple-inputs.csv"))
x1=[]
x2=[]
y=[]
for row in reader:
    x1.append(float(row[0]))
    x2.append(float(row[1]))
    y.append(float(row[2]))
lr=0.001
w1=2
w2=10
b=8
for epoch in range(100):
    for g in range(len(x1)):
        output=w1*x1[g]+w2*x2[g]+b*1
        error=y[g]-output
        gradient_w1=error*x1[g]
        gradient_w2=error*x2[g]
        gradient_b=error*1
        w1=w1+gradient_w1*lr
        w2=w2+gradient_w2*lr
        b=b+gradient_b*lr
print("final weight",w1,w2,b)
total_error=0
for i in range(len(x1)):
    output=w1*x1[i]+w2*x2[i]+b*1
    error=y[i]-output
    total_error+=error**2
print("MSE",total_error/len(x1))

```
