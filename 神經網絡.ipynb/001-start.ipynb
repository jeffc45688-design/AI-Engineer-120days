```python
import csv
reader=csv.reader(open("grades.csv"))
x=[]#輸入:時間
y=[]#輸出:成績
for row in reader:
    x.append(float(row[0]))
    y.append(float(row[1]))
# 建立模型 Y=W*X
lr=0.01#學習率 learning rate
w=2 #隨機給一個權重
for i in range(len(x)):
    output=w*x[i]
    error=y[i]-output
    gradient=error
    w=w+gradient*lr
print("Final Weight",w)
# 評估模型  MSE(Mean Square Error)均方誤差
total_error=0
for i in range(len(x)):
    output=w*x[i]
    error=y[i]-output
    total_error+=error**2
print("Mean Square Error",total_error/len(x))
```
