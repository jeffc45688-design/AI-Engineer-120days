```python
import csv
reader=csv.reader(open("grades-bias.csv"))
x=[]
y=[]
for row in reader:
    x.append(float(row[0]))
    y.append(float(row[1]))
# 建立模型 Y=W*X+B*1
lr=0.001 #給一個學習率
w=2     #隨機給一個權重
b=8     #隨機給一個權重
for epoch in range(2):
    for g in range(len(x)):
        output=w*x[g]+b*1
        error=y[g]-output
        gradient_w=error*x[g]
        gradient_b=error*1
        w=w+gradient_w*lr
        b=b+gradient_b*lr
    print("final weight",w,b)
total_error=0
for i in range(len(x)):
    output=w*x[i]+b*1
    error=y[i]-output
    total_error=total_error+error**2
print("MSE",total_error/len(x))
```
