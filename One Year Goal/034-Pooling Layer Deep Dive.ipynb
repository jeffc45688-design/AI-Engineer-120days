```python
import torch
import torch.nn as nn
X=torch.randn(1,8,26,26)
pool=nn.MaxPool2d(
    kernel_size=2,
    stride=2
)
Y=pool(X)
print(Y.shape)
print(X.shape)
```

```python
# MaxPool2d(2)=kneral_size=2 stride=2
# MaxPool2d(2,2)=kneral_size=2 stride=2
# MaxPool2d(3,2)=kneral_size=3 stride=2
# MaxPool2d(2,2,2)=kneral_size=2 stride=2 padding=2
```
