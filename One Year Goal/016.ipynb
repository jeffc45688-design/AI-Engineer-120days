```python
import torch 
import torch.nn as nn
import torch.optim as optim 
X=torch.tensor([
    [1.0,2.0],
    [2.0,1.0],
    [3.0,4.0],
    [4.0,3.0]
])
Y=torch.tensor([
    [0.0],
    [0.0],
    [1.0],
    [1.0]
])
class Net(nn.Module):
    def __init__(self):
        super().__init__()
        self.fc1=nn.Linear(2,4)
        self.fc2=nn.Linear(4,1)
    def forward(self,x):
        x=self.fc1(x)
        x=torch.relu(x)
        x=self.fc2(x)
        return x
model=Net()
criterion=nn.BCEWithLogitsLoss()
optimizer=optim.Adam(model.parameters(),lr=0.01)
epochs=500
for epoch in range(epochs):
    y_pred=model(X)
    loss=criterion(y_pred,Y)
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
    if epoch % 50 == 0:
        print(epoch,loss.item())
with torch.no_grad():
    pred=torch.sigmoid(model(X))
    label=(pred>0.5).int()
print("prediction",label.squeeze())
print("true",Y.squeeze())
#nn.MSELoss
#nn.BCEWithLogitsLoss
#nn.Module
#nn.Linear
```
