```python
import csv
import numpy as np
x1_list=[]
x2_list=[]
y_list=[]
with open("001.csv") as f:
    reader=csv.reader(f)
    for row in reader:
        x1_list.append(float(row[0]))
        x2_list.append(float(row[1]))
        y_list.append(float(row[2]))
x1=np.array(x1_list)
x2=np.array(x2_list)
y=np.array(y_list)
# print(y)
lr=0.1
b=1
X=np.column_stack((x1,x2))#(n,2)
Y=y.reshape(-1)#(n,)
W=np.array([2,8])#(2,)
n=len(X)
for epoch in range(100):
    output=X@W+b
    error=Y-output#(n,)加一維度才能相乘
    gradient_W=(error[:,np.newaxis]*X).mean(axis=0)
    gradient_b=error.mean()
    W=W+gradient_W*lr
    b=b+gradient_b*lr
output=X@W+b
mse=np.mean((Y-output)**2)    
print("final weight",W,b)
print("MSE",mse)




```
