```python
text="i love ai"
tokens=text.split()
print(tokens)
vocab={}
for i,word in enumerate(tokens):
    vocab[word]=i
vocab["<UNK>"]=len(vocab)
print("vocab",vocab)
text2="i love gpt"
tokens2=text2.split()
numbers=[vocab.get(word,vocab["<UNK>"]) for word in tokens2] #dict.get(key,default) key存在回傳value key不在回傳default
print("numbers",numbers)
```
