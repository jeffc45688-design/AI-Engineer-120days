```python
import numpy as np
X=np.array([
    [1,2],
    [2,1],
    [3,4],
    [4,3]
])#(4,2)
W=np.array([0.3,-0.2])#(2,)
b=0.1
z=X@W+b
print("z:",z)#(4,)矩陣@向量 結果還是像向量
def sigmoid(x):
    return 1/(1+np.exp(-x))
a=sigmoid(z)
print("sigmoil output",a)
W1=np.random.randn(2,3)*0.1#(2,3) 
# randn → 取 標準常態分布（mean=0，std=1）
# 值有正有負
# 非常適合用來當神經網路初始權重
b1=np.zeros(3)#(3,)
hidden=X@W1+b1
hidden_act=sigmoid(hidden)
print("hidden layer output",hidden_act)
```
