```python
import numpy as np 
X=np.array([
    [1,2],
    [2,1],
    [3,4],
    [4,3]
])#(4,2)
Y=np.array([0,0,1,1])#(4,)
def sigmoid(z):
    return 1/(1+np.exp(-z))
w1=np.random.randn(2,3)#(2,3)
b1=np.zeros(3)#(3,)
w2=np.random.randn(3)#(3,)
b2=0
lr=0.1
eps=1e-8
n=len(X)
epochs=1000
for epoch in range(epochs):
    z1=X@w1+b1
    a1=sigmoid(z1)
    z2=a1@w2+b2
    y_pred=sigmoid(z2)
    loss=-np.mean(Y*(np.log(y_pred+eps))+(1-Y)*(np.log(1-y_pred+eps)))
    gra_z2=y_pred-Y
    gra_z2_col=gra_z2[:,None]
    gra_z1=gra_z2_col*w2*a1*(1-a1)
    gradient_w1=(X.T@gra_z1)/n
    gradient_b1=gra_z1.mean(axis=0)
    gradient_w2=(a1.T@gra_z2)/n
    gradient_b2=gra_z2.mean()
    w1-=gradient_w1*lr
    b1-=gradient_b1*lr
    w2-=gradient_w2*lr
    b2-=gradient_b2*lr
    if epoch % 100 == 0:
        print(epoch,loss)
print("W1",w1)
print("W2",w2)
print("B1",b1)
print("B2",b2)
z1=X@w1+b1
a1=sigmoid(z1)
z2=a1@w2+b2
y_pred=sigmoid(z2)
print("y_pred",y_pred)
y_lable_pred=(y_pred>=0.5).astype(int)
print("預測",y_lable_pred)
print("真實",Y)
```
