```python
import torch 
import torch.nn as nn
X=torch.randn(1,1,28,28) #X得圖片channel(1,3)和經過CNN後的in_channels不一樣
class CNN(nn.Module):
    def __init__(self):
        super().__init__()
        self.conv1=nn.Conv2d(
            in_channels=1,
            out_channels=4,
            kernel_size=3
        ) #(28-3+2*0)/1+1=26 (1,4,26,26)  #output=(input-kernel+2*p)/S+1
        self.conv2=nn.Conv2d(
            in_channels=4,
            out_channels=2,
            kernel_size=5
        )#(26-5+2*0)/1+1=22 (1,2,22,22)
        self.pool=nn.MaxPool2d(2)
        self.fc=nn.Linear(2*11*11,1)
    def forward(self,x):
        x=self.conv1(x)
        x=torch.relu(x)
        x=self.conv2(x)
        x=torch.relu(x)
        x=self.pool(x)
        x=x.reshape(x.size(0),-1)
        x=self.fc(x)
        return x
model=CNN()
Y=model(X)
print(X.shape)
print(Y.shape)
print(Y)
```
