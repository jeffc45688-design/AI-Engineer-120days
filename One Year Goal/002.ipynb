```python
import csv
import numpy as np
x1_list=[]
x2_list=[]
y_list=[]
with open("001.csv") as f:
    reader=csv.reader(f)
    for row in reader:
        x1_list.append(float(row[0]))
        x2_list.append(float(row[1]))
        y_list.append(float(row[2]))
x1=np.array(x1_list)
x2=np.array(x2_list)
y=np.array(y_list)
X=np.column_stack((x1,x2))#(50,2)
Y=y.reshape(-1)#(50,)
W=np.array([2,8])
b=1
lr=0.01
n=len(X)
for epoch in range(1000):
    output=X@W+b
    error=Y-output#(50,)
    # print(error[:5])#負數
    gradient_W=(X.T@error)/n
    gradient_b=error.mean()
    W=W+gradient_W*lr
    b=b+gradient_b*lr
output=X@W+b
mse=np.mean((Y-output)**2)
print("final weight",b,W)
print("MSE",mse)
```
