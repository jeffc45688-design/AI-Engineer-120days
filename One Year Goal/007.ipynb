```python
import numpy as np
np.random.seed(42)
X=np.array([
    [1,2],
    [2,1],
    [3,4],
    [4,3]
])
y_lable=np.array([0,1,2,1])
Y=np.zeros((4,3))
Y[np.arange(4),y_lable]=1
def relu(z):
    return np.maximum(0,z)
def softmax(z):
    exp_z=np.exp(z-np.max(z,axis=1,keepdims=True))
    return exp_z/np.sum(exp_z,axis=1,keepdims=True)
w1=np.random.randn(2,5)*np.sqrt(2/2)
b1=np.zeros((1,5))
w2=np.random.randn(5,3)*np.sqrt(2/5)
b2=np.zeros((1,3))
def loss_fn(Y,y_pred,eps=1e-8):
    return -np.mean(np.sum(Y*np.log(y_pred+eps),axis=1))
lr=0.1
epochs=2000
for epoch in range(epochs):
    z1=X@w1+b1
    a1=relu(z1)
    z2=a1@w2+b2
    a2=softmax(z2)
    loss=loss_fn(Y,a2)
    dz2=a2-Y
    dw2=a1.T@dz2/len(X)
    db2=dz2.mean(axis=0,keepdims=True)
    da1=dz2@w2.T
    dz1=da1.copy()
    dz1[z1<=0]=0
    dw1=X.T@dz1/len(X)
    db1=dz1.mean(axis=0,keepdims=True)
    w2-=lr*dw2
    b2-=lr*db2
    w1-=lr*dw1
    b1-=lr*db1
    if epoch % 200 == 0:
        print(f"Epoch{epoch},Loss{loss}")
print("最後LOSS",loss)
y_lable_pred=np.argmax(a2,axis=1)
print("預測",y_lable_pred)
print("實際",y_lable)
    

```
