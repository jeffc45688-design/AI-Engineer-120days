```python
import numpy as np
X=np.array([
    [1,2],
    [2,1],
    [3,4],
    [4,3]
])
Y=np.array([[0],[0],[1],[1]])
def sigmoid(z):
    return 1/(1+np.exp(-z))
def relu(z):
    return np.maximum(0,z)
def relu_back(dA,Z):
    dZ=dA.copy()
    dZ[Z <= 0]=0
    return dZ
np.random.seed(42)
w1=np.random.randn(2,3)*np.sqrt(2/2)
b1=np.zeros((1,3))
w2=np.random.randn(3,3)*np.sqrt(2/3)
b2=np.zeros((1,3))
w3=np.random.randn(3,1)*np.sqrt(2/3)
b3=np.zeros((1,1))

lr=0.1
n=len(X)
eps=1e-8
epochs=2000
for epoch in range(epochs):
    Z1=X@w1+b1
    A1=relu(Z1)
    Z2=A1@w2+b2
    A2=relu(Z2)
    Z3=A2@w3+b3
    A3=sigmoid(Z3)
    Y_pred=A3
    loss=-np.mean(Y*(np.log(Y_pred+eps))+(1-Y)*(np.log(1-Y_pred+eps)))
    dZ3=A3-Y
    dw3=A2.T@dZ3/n
    db3=dZ3.mean(axis=0,keepdims=True)
    dA2=dZ3@w3.T
    dZ2=relu_back(dA2,Z2)
    dw2=A1.T@dZ2/n
    db2=dZ2.mean(axis=0,keepdims=True)
    dA1=dZ2@w2.T
    dZ1=relu_back(dA1,Z1)
    dw1=X.T@dZ1/n
    db1=dZ1.mean(axis=0,keepdims=True)
    w3-=lr*dw3
    b3-=lr*db3
    w2-=lr*dw2
    b2-=lr*db2
    w1-=lr*dw1
    b1-=lr*db1
    if epoch % 200 == 0:
        print(f"Epoch{epoch},Loss{loss}")
print("訓練結束")
print("最後loss:",loss)
print("最後機率",A3)
Y_lable_pred=(A3 >= 0.5).astype(int)
print("預測",Y_lable_pred.ravel())
print(Y.ravel())
```

```python
import numpy as np
X=np.array([
    [1,2],
    [2,1],
    [3,4],
    [4,3]
])
Y=np.array([[0],[0],[1],[1]])
def sigmoid(z):
    return 1/(1+np.exp(-z))
def relu(z):
    return np.maximum(0,z)
def relu_back(dA,Z):
    dZ=dA.copy()
    dZ[Z<=0]=0
    return dZ
np.random.seed(42)
w1=np.random.randn(2,3)*np.sqrt(2/2)
b1=np.zeros((1,3))
w2=np.random.randn(3,3)*np.sqrt(2/3)
b2=np.zeros((1,3))
w3=np.random.randn(3,1)*np.sqrt(2/3)
b3=np.zeros((1,1))
lr=0.1
eps=1e-8
n=len(X)
epochs=2000
for epoch in range(epochs):
    z1=X@w1+b1
    a1=relu(z1)
    z2=a1@w2+b2
    a2=relu(z2)
    z3=a2@w3+b3
    a3=sigmoid(z3)
    loss=-np.mean(Y*(np.log(a3+eps))+(1-Y)*(np.log(1-a3+eps)))
    dz3=a3-Y#誤差
    dw3=a2.T@dz3/n
    db3=dz3.mean(axis=0,keepdims=True)  #dz每層的誤差 da往前傳遞的梯度訊號 用來算上一層的dz
    da2=dz3@w3.T
    dz2=relu_back(da2,z2)
    dw2=a1.T@dz2/n
    db2=dz2.mean(axis=0,keepdims=True)

    db1=dz1.mean(axis=0,keepdims=True)
    if epoch % 200 == 0:
        print(f"epoch{epoch},loss{loss}")
    w1-=lr*dw1
    b1-=lr*db1
    w2-=lr*dw2
    b2-=lr*db2
    w3-=lr*dw3
    b3-=lr*db3
print('最後的loss:',loss)
print("最後機率",a3)
Y_lable_pred=(a3>=0.5).astype(int)
print("預測",Y_lable_pred.ravel())
print("真實",Y.ravel())
```
