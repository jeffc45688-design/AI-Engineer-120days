```python
import torch 
import torch.nn as nn
seq_len=3 #幾個token
batch_size=1 #一次丟一句
input_size=4 #維度
hidden_size=5 #一步hidden state是5維
X=torch.randn(seq_len,batch_size,input_size)
rnn=nn.RNN(
    input_size=input_size,
    hidden_size=hidden_size,
    batch_first=False
)
output,hn=rnn(X)
print("X",X.shape)
print("output",output.shape)# [seq_len,batch_size,hidden_size]
print("hn",hn.shape)#[num_layers,batch_size,hidden_size]
```
