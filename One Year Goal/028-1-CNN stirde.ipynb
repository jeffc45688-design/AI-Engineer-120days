```python
import torch
import torch.nn as nn
X=torch.randn(1,1,6,6)
class CNN(nn.Module):
    def __init__(self):
        super().__init__()
        self.conv1=nn.Conv2d(1,2,3,stride=2)
        self.fc=nn.Linear(8,1)
    def forward(self,x):
        x=self.conv1(x)
        x=torch.relu(x)
        x=x.view(x.size(0),-1)
        x=self.fc(x)
        return x
model=CNN()
y=model(X)
print(X.shape)
print(y.shape)
print(y)
```
