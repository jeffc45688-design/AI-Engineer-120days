```python
import torch
import torch.nn as nn
X=torch.tensor([
    [1.0,2.0],
    [2.0,1.0],
    [3.0,4.0],
    [4.0,3.0]
])
class Net(nn.Module):
    def __init__(self):
        super().__init__()
        self.fc1=nn.Linear(2,4)
        self.fc2=nn.Linear(4,1)
    def forward(self,X):
        X=self.fc1(X)
        X=torch.relu(X)
        X=self.fc2(X)
        return X
model=Net()
model.load_state_dict(torch.load("best_model.pth"))
model.eval()
with torch.no_grad():
    logits=model(X)
    prob=torch.sigmoid(logits)
    label=(prob>0.5).float()
print("prediction",label.squeeze())
```
