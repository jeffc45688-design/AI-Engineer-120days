```python
# exp 是 exponential（指數）的縮寫，意思是：
# np.exp(x) = e 的 x 次方 = e^x
# 而 e = 2.718281828… 是一個固定的數（像 π=3.14）。
# 你可以把 e 想成：
# 每次成長 “2.718 倍” 的成長率
# 所以：
# exp(1) ＝ 變成 2.718 倍
# exp(2) ＝ 變成 2.718 的 2 次方（變超快）
# exp(-1) = 變成 1/2.718 倍（變小）
# np.exp(0) = e^0 = 1
# 任何數的 0 次方都等於 1
```

```python
import numpy as np
X=np.array([
    [1,2],
    [2,1],
    [3,4],
    [4,3]
])#(4,2)
Y=np.array([0,0,1,1,])#(4,)
def sigmoid(z):
    return 1/(1+np.exp(-z))
n,d=X.shape
W=np.zeros(d)
b=0
lr=0.1
num_epochs=1000
for epoch in range(num_epochs):
    z=X@W+b
    y_pred=sigmoid(z)
    eps=1e-8
    loss=-np.mean(Y*(np.log(y_pred+eps))+(1-Y)*(np.log(1-y_pred+eps)))
    error=y_pred-Y
    gradinet_W=X.T@error/n
    gradinet_b=error.mean()
    W-=gradinet_W*lr
    b-=gradinet_b*lr
    if epoch % 100==0:
        print(f"epoch={epoch},loss={loss}")
print("final Weight",W,b)
z=X@W+b
y_pred=sigmoid(z)
print("y_pred",y_pred)
y_label_pred=(y_pred>=0.5).astype(int)
print("y_label_pred",y_label_pred)
print("Y true",Y)
accuracy=np.mean(y_label_pred==Y)
print("Accuracy",accuracy)
```
