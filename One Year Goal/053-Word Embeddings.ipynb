```python
import torch
import torch.nn as nn
vocab={"i":0,"love":1,"ai":2,"<UNK>":3}
tokens=["ai","love","ai"]
numbers=[vocab[word] for word in tokens]
print(numbers)
input_tensor=torch.tensor(numbers)
print(input_tensor)
embedding=nn.Embedding(
    num_embeddings=len(vocab),
    embedding_dim=4
)
output=embedding(input_tensor)
print(output)
print(output.shape)
```
