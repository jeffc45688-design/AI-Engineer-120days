```python
def test(arg):
    print(arg)
#回呼函式
def test2():
    print(100)

test(test2)
```

```python
def test(arg):
    arg()#呼叫回呼函式

def test2():
    print(100)

test(test2)
```

```python
def test(arg):
    arg("Hi")#呼叫回呼函式,代入參數

def test2(da):
    print(da)

test(test2)
```
