```python
#載入pandas模組
import pandas as pd
#建立Series(類似列表list)
data=pd.Series([20,10,15])
# print(data)
# print("MAX",data.max())
# print("Median",data.median())
# data=data*2
# print(data)
data=data==20 #布林遮罩（Boolean Masking)逐個元素比較，產生布林值
print(data)
```
