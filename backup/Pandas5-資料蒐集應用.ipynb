```python
import pandas as pd
data=pd.read_csv("googleplaystore.csv")#把CSV格式的檔案讀取成一個DataFrame檔案
# print("資料數量",data.shape)
print("資料欄位",data.columns)
condition=data["Rating"]>5
data=data[condition]
print(data)
print(data["Rating"])
print("平均數",data["Rating"].mean())
print("取得前100應用程式的平均",data["Rating"].nlargest(100).mean()
      )
```

```python
import pandas as pd
data=pd.read_csv("googleplaystore.csv")
# 排除奇怪資料
condition=data["Rating"]<=5
data=data[condition]
print(data["Rating"])
print("平均數",data["Rating"].mean())
print("取得前1000應用程式的平均",data["Rating"].nlargest(1000).mean()
      )
```

```python
import pandas as pd
data=pd.read_csv("googleplaystore.csv")
data["Installs"]=pd.to_numeric(data["Installs"]
    .str.replace("[,+]","",regex=True)
    .replace("Free",None))#Free 變成 None → to_numeric 可解析
     ## 若還有奇怪值 → 轉 NaN 不會報錯
# None / NaN 可以被 pd.to_numeric 正確處理
# print(data["Installs"])
print("平均數",data["Installs"].mean())
condition=data["Installs"]>100000
print("安裝數量大於十萬",data[condition].shape[0])
```

```python
data["Installs"] = pd.to_numeric(
    data["Installs"]
        .astype(str)
        .str.replace(r"[,+]", "", regex=True)
        .str.replace("Free", "", regex=False),
    errors="coerce"
)
```
