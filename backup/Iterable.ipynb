```python
#可疊代Iterable資料:字串 列表 集合 字典
# for 變數名稱 in 可疊代資料:
for x in "abc":
    print(x)
for x in {"a":"1","b":"2"}:#字典針對的是KEY的部分
    print(x)
dic={"a":"1","b":"2"}#拿出Value
for x in dic:
    print(dic[x])
# 內建函式
# max
result=max([10,2,30,1])
print(result)
result=max("xyz")
print(result)
result=max({10,2000,30,1})
print(result)
result=max({"x":3,"a":4})#只看KEY
print(result)
#sorted
result=sorted("cba")
print(result)#轉成列表小到大
```
