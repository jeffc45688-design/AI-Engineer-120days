```python
import pandas as pd
# 資料索引:pd.DataFrame(字典,index=索引列表)
data=pd.DataFrame({
    "name":["Amy","Bob","Charles"],
    "salary":[30000,50000,40000]
},index=["a","b","c"])
print(data)
```

```python
#觀察資料
import pandas as pd
data=pd.DataFrame({
    "name":["Amy","Bob","Charles"],
    "salary":[30000,50000,40000]
},index=["a","b","c"])
print("資料數量",data.size)
print("資料形狀(列,欄)",data.shape)
print("資料索引",data.index)
```

```python
# 取得列(Row/橫向)的Series的資料,根據順序,索引
import pandas as pd
data=pd.DataFrame({
    "name":["Amy","Bob","Charles"],
    "salary":[30000,50000,40000]
},index=["a","b","c"])
print("取得第二列",data.iloc[1],sep="\n")
#sep="\n"每個項目換行
print("取得c列/直向",data.loc["c"],sep="\n")
```

```python
# 取得欄(Column/直向)的Series的資料,根據順序,索引
import pandas as pd
data=pd.DataFrame({
    "name":["Amy","Bob","Charles"],
    "salary":[30000,50000,40000]
},index=["a","b","c"])
print("取得name欄位",data["name"],sep="\n")
#單獨取得列或欄會從雙維度變成Series單維度
names=data["name"]
print("把name全部換成大寫",names.str.upper(),sep="\n")
salaries=data["salary"]
print("薪水的平均值:",salaries.mean())
```

```python
# 建立新的欄位
import pandas as pd
data=pd.DataFrame({
    "name":["Amy","Bob","Charles"],
    "salary":[30000,50000,40000]
},index=["a","b","c"])
data["revenue"]=[500000,400000,300000]#加入欄位
# print(data)
data["rank"]=pd.Series([1,2,3],index=["a","b","c"])#加入新欄位(正式),新欄位等於加一個Series進去
#前面有index所以這也要加 不然無法顯示
data["cp"]=data["revenue"]/data["salary"]
print(data)
```
