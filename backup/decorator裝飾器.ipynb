```python
# 特殊設計的"函式" 用來"輔助"其他函式
def 裝飾器名稱(回呼函式名稱):
    def 內部函式名稱():
        回呼函式名稱():

    return 內部函式名稱

```

```python
def myDeco(cb):
    def run():
        print("裝飾器中的程式碼")
        cb(5)#這個回呼函式,就是被裝飾得普通函式
    return run
@myDeco
def test(n):
    print("普通程式碼",n)
test()
```

```python
# 定義一個裝飾器 計算+1+2...+50的總和
def calculate(callback):
    def run():
        sum=0
        for x in range(51):
            sum=sum+x
        print(sum)
        callback()
    return run
@calculate
def show():
    print("普通程式的程式碼")

show()
```

```python
def calculate(callback):
    def run():
        sum=0
        for x in range(51):
            sum=sum+x
     
        callback(sum)
    return run
@calculate
def show(n):
    print("結果是",n)
@calculate
def showEN(n):
    print("The result is",n)
show()
showEN()
# 重複利用裝飾器內的程式碼得到不同的表示方式
```
