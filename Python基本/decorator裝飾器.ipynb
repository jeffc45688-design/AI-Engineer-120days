```python
def 裝飾器名稱(回呼函式名稱):
    def 內部函式名稱():

        回呼函式名稱()
    return 內部函式名稱 
```

```python
def myDeco(callback):
    def run():
        print("裝飾器中的程式碼")
        callback(5)
    return run
@myDeco
def test(n):
    print("普通函式的程式碼",n)
test()
    
        
```

```python
def calculate(callback):
    def run():
        sum=0
        for x in range(51):
            sum=sum+x
        callback(sum)
    return run#沒小括號
@calculate#裝飾器不用小括號
def show(n):
    print("總和是",n)
@calculate
def show1(n):
    print("The total is",n)
show()
show1()
```
