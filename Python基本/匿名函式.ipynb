```python
# lambda function匿名函式
def multiply(n1,n2):
    return n1*n2
result=multiply(3,4)
print(result)

mul=lambda n1,n2:n1*n2
result=mul(3,4)
print(result)
```

```python
from datetime import datetime
def now():
    return datetime.now()
print(now())

now=lambda :datetime.now()#lambda 參數: 運算式
print(now())
```

```python
ages=[5,20,18,3,80,10]
a=filter(lambda a:a>=18,ages)
a=list(a)
print(a)
```

```python
prices=[100,50,200,75] #MAP內建映射函式
DisPrices=map(lambda p:p*0.8,prices)
print(list(DisPrices))
```
