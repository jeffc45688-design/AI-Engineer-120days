datetime 模組
│
├── date          （只有日期）
├── time          （只有時間）
├── datetime      （日期＋時間）
├── timedelta     （時間差）
├── timezone      （固定時區物件）
└── tzinfo        （時區抽象基底類別）

```python
from datetime import date, timedelta
dt=date.today()
print(dt.year,dt.month,dt.day)
print(dt)
```

```python
from datetime import date, timedelta
dt=date(2000,1,1)
print(dt)
print(dt.year,dt.month,dt.day)
print(dt.weekday())#星期幾 從0開始
```

```python
from datetime import date, timedelta
dt=date.today()
diff=timedelta(weeks=4)
afterWeekdate=dt+diff
print(afterWeekdate)
```

```python
from datetime import date, timedelta
weekday_names={0:"週一",1:"週二",2:"週三",3:"週四",4:"週五",5:"週六",6:"週日"}
today=date.today()
dt=date(today.year,today.month,1)
while dt.month==today.month:
    weekday=dt.weekday()
    print(dt,weekday_names[weekday])
    dt=dt+timedelta(days=1)
```
