```python
#儲存檔案
file=open("data.txt",mode="w",encoding="utf-8")#開啟檔案
file.write("測試中文\n好棒棒")#操作 #enconding="utf-8" 才能用中文
file.close()#關閉檔案 
#最佳實務
with open("data.txt",mode="w",encoding="utf-8")as file:
    file.write("5\n3\n4")
```

```python
#讀取檔案
#把檔案的數字資料，一行一行的讀取出來並計算總合
sum=0
with open("data.txt",mode="r",encoding="utf-8")as file:
    for x in file:#一行一行的讀取
        sum+=int(x)
print(sum)
```

```python
# 使用JSON格式讀取,複寫檔案
import json
with open("config.json",mode="r") as file:
    data=json.load(file)
print(data)#data是一個字典資料
# print("name:",data["name"])
# print("version:",data["version"])
#修改資料
data["name"]="New Name"#修改變數資料
#把最新資料複寫回檔案中
with open("config.json",mode="w") as file:
   json.dump(data,file) #把data的資料寫入至變數file裡
print(data)




```
