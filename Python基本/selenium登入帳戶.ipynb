```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.keys import Keys
import time
options=Options()
driver=webdriver.Chrome()
driver.get(r"https://leetcode.com/accounts/login/?next=%2Fproblemset%2F")
usernameInput=driver.find_element(By.ID,"id_login")
passwordInput=driver.find_element(By.ID,"id_password")
usernameInput.send_keys("帳號")
passwordInput.send_keys("密碼")
signinBtn=driver.find_element(By.ID,"signin_btn")
signinBtn.send_keys(Keys.ENTER)
time.sleep(5)
statElement=driver.find_element(By.CSS_SELECTOR,"")
columns=statElement.text.split("\n")
print("columns")
driver.quit
```

CSS_SELECTOR 是一種用 CSS 語法定位網頁元素的方法。
能精準選到 ID、Class、Tag、階層、屬性… 幾乎任何元素。

方法	能力	是否精準	是否推薦
By.ID	用 ID	⭐⭐⭐⭐⭐	✔ 很準但網站常換
By.CLASS_NAME	Class	⭐⭐⭐	❌ 一定會撞到很多同名 class
By.XPATH	強大	⭐⭐⭐⭐⭐	✔ 但太長、太醜
By.CSS_SELECTOR	最通用	⭐⭐⭐⭐⭐	✔✔✔ 建議主力使用

用 ID 找元素（最簡單）
# <input id="id_login">
driver.find_element(By.CSS_SELECTOR, "#id_login")
# = by id

用 Class 找元素
# <div class="base-search-card__title">
driver.find_element(By.CSS_SELECTOR, ".base-search-card__title")
. = by class
注意 class 名稱有底線也完全 OK。

找一個 tag
driver.find_element(By.CSS_SELECTOR, "button")
driver.find_element(By.CSS_SELECTOR, "input")
driver.find_element(By.CSS_SELECTOR, "h1")
