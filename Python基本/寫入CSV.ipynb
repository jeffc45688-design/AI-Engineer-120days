```python
import csv
with open("data.csv",mode="w",newline="") as file:#避免 Windows 多出空白行（CSV 官方建議）
    writer=csv.writer(file)
    writer.writerow([1,2,3])
    writer.writerow([4,5,6])
    writer.writerow([7,8,9])
```

```python
import csv
with open("data.csv",mode="r",newline="") as file:
    reader=csv.reader(file)
    total=0
    for row in reader:#逐列讀取資料
        for nb in row: #nb是字串
            total=total+int(nb)#轉成數字後相加
    print(total)
```
