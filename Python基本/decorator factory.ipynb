```python
def myFactory(base):#裝飾器工廠能加入額外參數,裝飾器不行,是回呼函式
    def myDeco(callback):
        def run():
            print("裝飾器內的程式",base)
            callback()
        return run#沒小括號
    return myDeco#沒小括號
@myFactory(3)#要小括號
def test():
    print("普通函式的程式")
test()
```

```python
def myFactory(base):
    def myDeco(callback):
        def run():
            print("裝飾器內的程式",base)
            result=base*2
            callback(result)
        return run
    return myDeco
@myFactory(3)
def test(n):
    print("普通函式的程式",n)
test()
```

```python
#計算+1+2...+n
def myFac(g):
    def myDec(cb):
        def run():
            sum=0
            for x in range(g+1):
                sum+=x
            cb(sum)
        return run
    return myDec
@myFac(100)
def answer(n):
    print("總和是",n)
@myFac(50)
def answer1(n):
    print("The tatal is",n)
answer()
answer1()
```
