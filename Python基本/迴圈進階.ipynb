```python
# break範例
n=0
while n<5:
    if n==3:
        break
    print(n)#迴圈內的n
    n+=1
print("最後的",n)#迴圈結束的n

```

```python
#continue範例
n=0
for x in[0,1,2,3]:      
    if x%2==0:  # x是偶數得意思 % = 取餘數
        continue
    print(x)
    n+=1
print("最後的n:",n)  #n+=1被執行2次 所以等於2
```

```python
#else的範例
sum=0
for x in range(11):  #0-10
    sum+=x
else:
    print(sum) #印出0+1+2..+10的結果 else:迴圈結束前,先跑此區塊命令
```

```python
for x in [1,3,5]:
    if x == 4:
        print("找到 4 了！")
        break
else:
    print("找不到 4")  #for else搭配break
```

```python
#綜合範例:找出整數平方根
#輸入9得到3
#輸入11得到[沒有]平方根
n=input("輸入一個正整數:")
n=int(n) # 轉換成數字
for x in range(n):  # x 從 0 ~ n-1
    if x*x==n:
        print("整數平方根",x)
        break #break強制結束時就不會跑else
else:
    print("沒有整數平方根")
```
