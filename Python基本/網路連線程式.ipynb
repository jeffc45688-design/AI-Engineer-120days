```python
#網路連線
import urllib.request as  request

import ssl
context = ssl._create_unverified_context()

src="https://www.ntu.edu.tw/"

with request.urlopen(src, context=context) as response:
        data=response.read().decode("utf-8") #取得台灣大學網站原始碼(html.css.js)
print(data)
```

```python
import urllib.request as  request
import json
import ssl#SSL 驗證被擋住
context = ssl._create_unverified_context()

src="https://data.taipei/api/v1/dataset/25edd484-210a-43d7-962c-9df37be4b11d?scope=resourceAquire"
with request.urlopen(src,context=context) as response:
        data=json.load(response)
print(data)
```

```python
import ssl
ssl._create_default_https_context = ssl._create_unverified_context 

import urllib.request as  request
import json
src="https://data.taipei/api/v1/dataset/25edd484-210a-43d7-962c-9df37be4b11d?scope=resourceAquire"
with request.urlopen(src) as response:
        data=json.load(response)
print(data)
clist=data["result"]["results"]
with open("data.txt","w",encoding="utf-8")as file:
      for company in clist:
        file.write(company["制定單位"]+"\n")



```
