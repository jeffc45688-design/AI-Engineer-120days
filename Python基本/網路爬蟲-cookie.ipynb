```python
#抓取PTT八卦版得網頁原始碼(HTML)
import urllib.request as req
def getData(url):#主程序:抓取多頁面標題
    request=req.Request(url,headers={"cookie":"over18=1","User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/144.0.0.0 Safari/537.36"})
    with req.urlopen(request) as response:
        data=response.read().decode("utf-8")
    import bs4
    root=bs4.BeautifulSoup(data,"html.parser")
    titles=root.find_all("div",class_="title")
    for title in titles:
        if title.a:
            print(title.a.string)
    nextlink=root.find("a",string="‹ 上頁")#抓取上一頁網址
    return nextlink["href"]
#抓取一個頁面的標題
pageurl="https://www.ptt.cc/bbs/Gossiping/index.html"
count=0
while count<5:#一次抓5頁
    pageurl="https://www.ptt.cc"+getData(pageurl)
    count+=1
```
