```python
from datetime import datetime
now=datetime.now()
print(now)
print(now.year,now.month,now.day)
print(now.hour,now.minute,now.second)
print(now.weekday())
time=datetime(2000,1,1,1,1,1)
print(time)
```

```python
from datetime import timedelta,timezone,datetime
now=datetime.now()
diff=timedelta(hours=8,minutes=30)
newtime=now+diff
print(newtime)
```

```python
from datetime import timedelta,timezone,datetime
jp_tz=timezone(timedelta(hours=9))
jp_now=datetime.now(tz=jp_tz)
print(jp_now)
t=datetime(2025,3,5,16,16,20)
jp_t=t.astimezone(tz=jp_tz)
print(jp_t)
```
