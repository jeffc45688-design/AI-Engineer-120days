```python
#例外處理情境:轉換資料型態失敗

data=input("請輸入一個數字")
try:
    number=int(data)
except Exception:
    number=0
number=number*2
print(number)
```

```python
#輸入資料不是數字,重新輸入
while True:
    data=input("請輸入一個數字")
    try:
        number=int(data)
        break
    except Exception:
        print("格式錯誤,請重新輸入一個數字")
number=number*2
print(number)
```
