```python
#函式def(define)#回傳值return
#定義函式
#函式內部的程式碼，若是沒有做函式呼叫，則不會被執行
def multiply(n1,n2):
    print(n1*n2)
#呼叫函式
multiply(3,4)
multiply(5,6)
#範例2 #有回傳值的函式  
def multiply2(n1,n2):
    print(n1*n2)
    return n1*n2
result = multiply2(7,8)
print("回傳值:",result)
#範例3
def multiply3(n1,n2):
    return n1*n2
result2 = multiply3(9,10)+multiply3(11,12)
print("回傳值相加:",result2)
#範例4  add函式沒有回傳值，預設回傳None
def add(n1,n2):
    print(n1+n2)
result3 = add(1,2)
print("回傳值:",result3)
#程式的包裝 
#函式可用來做程式的包裝，同樣的邏輯可重複利用
def calculate(max):
    sum=0
    for x in range(1,max+1):
         sum += x
    print(sum)
calculate(10)
calculate(20)
```
