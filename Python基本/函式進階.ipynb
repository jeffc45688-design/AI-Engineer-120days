```python
# 參數的預設資料
def power(base,exp):
    print(base**exp)
power(3,2)
def power(base,exp=0):#預設指數為0
    print(base**exp)
power(4) #沒有給指數,但任何數0次方等於1
# base = 底數
# 你要被「次方」運算的數字。
# exp = 指數（exponent）
# 表示「要乘幾次」
```

```python
# 使用名稱對應
def divide(n1,n2):
    print(n1/n2)
divide(2,4)
divide(n2=2,n1=4)
```

```python
# 無限/不定參數資料
def avg(*n):
    print(n) 
avg(3,4)
avg(3,5,10)
avg(1,4,-1,-8) #轉換成tuple
def avg(*n):
    sum=0
    for x in n:
        sum+=x
    print(sum/len(n))
avg(3,4)
avg(3,5,10)
avg(1,4,-1,-8)
# len() 是 Python 的內建函式
# 用來取得：
# list 長度
# tuple 長度
# 字串長度
# len(n) 代表你傳了幾個數字進去
```
