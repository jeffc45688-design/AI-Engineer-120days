```python
#隨機模組
import random
#隨機選取
data=random.choice([1,5,6,10,20])
print(data)
data=random.sample([1,5,6,10,20],3)#隨機選三筆
print(data)
```

```python
#隨機調換順序
data=[1,5,8,20]
random.shuffle(data) #調換整個data(洗牌)
print(data)
```

```python
#  隨機取得的亂數
data=random.random()#0~1之間的隨機亂數
print(data)
data=random.uniform(60,100)
data=int(data)
print(data)
```

```python
# 常態分配亂數 
data=random.normalvariate(100,10)#平均100 標準差10
print(data)
```

```python
# 統計模組
import statistics as sta
data=sta.mean([1,4,5,8]) #maen平均數 median中位數(能去掉極端數)
print(data)
data=sta.median([1,2,3,4,5,8,100])
print(data)
data=sta.stdev([1,2,3,4,5,8,10])#標準差
print(data)
data=sta.mode([1,2,2,4,5,6])#眾數 注意：如果資料沒有重複值，mode 可能會報錯。
print(data)
```

功能	函式
平均數	statistics.mean()
中位數	statistics.median()
眾數	statistics.mode()
標準差	statistics.stdev()
變異數	statistics.variance()
幾何平均	statistics.geometric_mean()
Harmonic Mean	statistics.harmonic_mean()
