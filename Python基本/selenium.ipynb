```python
# selenium：一個可以用程式控制瀏覽器（Chrome/Edge/Firefox）的工具。
# webdriver：Selenium 用來控制瀏覽器的核心模組。
# Options：用來設定 Chrome 的啟動參數（例如無痕、無頭、語言、視窗大小等）。
# selenium/
#     webdriver/    # 這是一個資料夾（模組）
#         Chrome     # 這是裡面的類別
#         Firefox
#         Edge
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
#設定Chrome Driver的執行檔路徑
options=Options()
options.chrome_executable_path=r"C:\Users\marce\ai-career-2026\Python基本\chromedriver.exe"
# 建立Driver物件實體,用程式操作瀏覽器運作
driver=webdriver.Chrome(options=Options)
driver.maximize_window()#視窗最大化
driver.get("https://www.google.com/")
driver.save_screenshot("screenshot-google.png")
driver.close()
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options

options = Options()
service = Service(r"C:\Users\marce\ai-career-2026\Python基本\chromedriver.exe")

driver = webdriver.Chrome(service=service, options=options)
driver.maximize_window()
driver.get("https://www.google.com/")
driver.save_screenshot("screenshot-google.png")
driver.quit()
```
