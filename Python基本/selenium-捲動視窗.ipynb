```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
import time
options=Options()
driver=webdriver.Chrome()
driver.get("https://www.linkedin.com/jobs/search?trk=guest_homepage-basic_guest_nav_menu_jobs")
driver.execute_script("window.scrollTo(0,document.body.scrollHeight);")#捲動視窗至底部
time.sleep(3)#等待三秒
titleTags=driver.find_elements(By.CLASS_NAME,"base-search-card__title")
for titleTag in titleTags:
    print(titleTag.text)
driver.quit()
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
import time
options=Options()
driver=webdriver.Chrome()
driver.get("https://www.linkedin.com/jobs/search?trk=guest_homepage-basic_guest_nav_menu_jobs")
n=0
while n<3:
    driver.execute_script("window.scrollTo(0,document.body.scrollHeight);")#捲動視窗至底部
    time.sleep(3)#等待三秒
    n+=1
titleTags=driver.find_elements(By.CLASS_NAME,"base-search-card__title")
for titleTag in titleTags:
    print(titleTag.text)
driver.quit()
```
