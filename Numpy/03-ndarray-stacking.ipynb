```python
import numpy as np
arr1=np.array([
    [1,2,3],
    [4,5,6]
])#2x3
arr2=np.array([
    [7,8,9],
    [10,11,12]
])#2x3
arr3=np.array([
    [13,14],
    [15,16]
])#2x2
result=np.vstack((arr1,arr2))#欄數必須相同
# np.concatenate((a,b), axis=0)
# 垂直堆疊
# 上下接起來
# 列數會增加
print(result.shape)#4x3
print("--------------------------")
result2=np.hstack((arr1,arr2))#列數必須相同
# np.concatenate((a,b), axis=1
# 水平堆疊
# 左右接起來
# 欄數會增加
print(result2.shape)#2x6
```

shape = (列數, 欄數)
         ↑      ↑
       axis=0  axis=1
axis	意思	會影響哪個方向
axis=0	上下方向	列（row）
axis=1	左右方向	欄（column）
