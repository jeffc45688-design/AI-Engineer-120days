```python
import numpy as np
# 一維陣列
data=np.array([3,2,6,4])
print(data)
data=np.empty(4)#創造一個有四個資料的一維陣列,資料未指定
print(data)
data=np.zeros(3)
print(data)
data=np.ones(3)
print(data)
data=np.arange(5)
print(data)

```

```python
# 二維陣列
import numpy as np
data=np.array([ #3x3二維陣列
    [2,3,2],
    [1,5,2],
    [4,2,1]
])
print(data)
data=np.empty([3,3])#創造一個3x3的二維陣列,資料未指定
print(data)
data=np.ones([2,3])
print(data)
```

```python
# 三維陣列
data=np.array([  #2X2X2的三維陣列
    [
        [3,1],[1,2]
    ],
    [
        [2,5],[10,2]
    ]
])
print(data)
data=np.zeros([3,1,3])
print(data)
```

```python
# 高維陣列
data=np.array([#1X1X2X3的高維陣列
    [
        [
            [3,2,1],
            [5,5,10]
            
        ]
    ]
])
print(data)
data=np.ones([2,1,1,2])
print(data)
```
