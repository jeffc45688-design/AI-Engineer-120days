```python
import numpy as np
# 逐元運算(elementwise) #shape要一模一樣
data1=np.array([2,3,10])
data2=np.array([1,3,6])
result=data1+data2
print(result)
result=data1*data2
print(result)
result=data1>data2
print(result)
result=data1==data2
print("是否相等",result)
```

```python
# 矩陣運算(matrix) 矩陣 = 二維陣列（2D array）
data1=np.array([#1*2
    [1,3]
])                      #內積2需對上2才能算col=row
data2=np.array([#2*3
    [2,-1,3],
    [-2,4,1]
])                      #shape=1x3 
# result=data1.dot(data2)
result=data1@data2 #內積
print(result)
print("-------------------")
result=np.outer(data1,data2)#2(1*2)X6(2*3) 外積
print(result)
```
